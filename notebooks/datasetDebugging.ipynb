```python
import pandas as pd

# Read only the first few lines of the JSONL file
df = pd.read_json('../data/parsed/replays.jsonl', lines=True, nrows=5)
print(df.head())
print(f"\nDataFrame shape: {df.shape}")
print(f"\nColumn names: {df.columns.tolist()}")
```

```python
# Inspect the structure of each nested column
print("Meta fields:")
print(df['meta'].iloc[0])
print(f"Type: {type(df['meta'].iloc[0])}")

print("\n" + "="*50 + "\n")

print("Observer_state fields:")
print(df['observer_state'].iloc[0])
print(f"Type: {type(df['observer_state'].iloc[0])}")

print("\n" + "="*50 + "\n")

print("Opponent_state fields:")
print(df['opponent_state'].iloc[0])
print(f"Type: {type(df['opponent_state'].iloc[0])}")

print("\n" + "="*50 + "\n")

print("Actions fields:")
print(df['actions'].iloc[0])
print(f"Type: {type(df['actions'].iloc[0])}")

print("\n" + "="*50 + "\n")

print("Target fields:")
print(df['target'].iloc[0])
print(f"Type: {type(df['target'].iloc[0])}")
```
